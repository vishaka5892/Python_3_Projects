# Mortgage Calculator - 
Calculate the monthly payments of a fixed term mortgage over given Nth terms at a given interest rate. Also figure out how long it will take the user to pay back the loan. For added complexity, add an option for users to select the compounding interval (Monthly, Weekly, Daily, Continually).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_price = float(input("Enter the selling price of the house in dollars($) :"))
down_payment_percent = float(input("Enter the down payment percent e.g 3 for 3% : ")) 
loan_type = int(input("Enter the mortgage term (in years) e.g. 30 for 30 years : "))
#loan_total = float(input("Enter the amount of loan : "))
loan_to_be_paid = sales_price*(1-down_payment_percent/100)
loan_loan = sales_price*(1-down_payment_percent/100)
loan_interval = int(12*loan_type)
interest_rate = float(input("Enter the interest rate e.g 2 for 2%: "))

Rate = 1+(interest_rate)/(12*100)

Amount_A = loan_to_be_paid*(Rate**loan_interval)*(1-Rate)/(1-Rate**loan_interval)

monthly_interest = []
balance = []

for i in range(1,loan_interval+1):
    interest = loan_to_be_paid*(Rate-1)
    loan_to_be_paid = loan_to_be_paid - (Amount_A-interest)
    monthly_interest = np.append(monthly_interest,interest)
    balance = np.append(balance,loan_to_be_paid)
    

print("The Home Sales Price is: = " + str('$')+ str(sales_price))
print("The Down Payment as a Percentage of Sales Price is: = " + str(down_payment_percent)+str(' %'))
print("The Loan Amount is: = " + str(loan_loan)+str(' %'))
print("The Interest Rate on Annual Percentage Basis is: = " + str(Rate)+str(' %'))
print("The duration of this loan(in months) is: = " + str(loan_interval)+str(' months'))
print("Monthly Payment for this Mortgage(P & I) is: = " + str('$')+str(np.round(Amount_A,2)))
print("Total interest paid over life cycle of the loan is: = " + str('$') + str(np.round(np.sum(monthly_interest),2)))



```

```python
plt.plot(range(1,loan_interval+1),monthly_interest, 'r',lw=2)
plt.xlabel('month')
plt.ylabel('monthly interest ($)')
plt.show()
plt.plot(range(1,loan_interval+1),balance,'b',lw=2)
plt.xlabel('month')
plt.ylabel('monthly loan balance ($)')
plt.show()
```
